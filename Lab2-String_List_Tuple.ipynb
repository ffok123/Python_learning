## Escape Sequences

Back slashes represent the beginning  of escape sequences. Escape sequences represent strings that may be difficult to input. For example, back slash "n" represents a new line. The output is given by a new line after the back slash "n" is encountered:


```python
# New line escape sequence

print(" Harry Potter \n is the best" )
```

Similarly, back slash  "t" represents a tab:


```python
# Tab escape sequence

print(" Harry Potter \t is the best" )
```

If you want to place a back slash in your string, use a double back slash:


```python
# Include back slash in string

print(" Harry Potter \\ is the best" )
```

We can also place an "r" before the string to display the backslash:


```python
# r will tell python that string will be display as raw string

print(r" Harry Potter \ is the best" )
```

## List Indexing

We can use the index operator [] to access an item in a list. In Python, indices start at 0. So, a list having 5 elements will have an index from 0 to 4.

```python
my_list = ["Apple","Orange","Lemon","Watermelon"]

# first item
print(my_list[0])  

# third item
print(my_list[2])  

# Print the elements on each index

print('print the first element using postive index:',my_list[0],
'\nprint the first element using negative index:' , my_list[-4]  )
```

```python
# fifth item
print(my_list[4])  # error
```

Lists can contain strings, floats, and integers. We can nest other lists, and other data structures.

```python
# Nested List
n_list = ["Python", [2, 0, 1, 5]]

# Nested indexing
print(n_list[0][1])

print(n_list[1][3])
```

## List slicing

```python
# List slicing in Python

my_list = ['p','y','t','h','o','n']

# elements from index 2 to index 4
print(my_list[2:5])

# elements from index 5 to end
print(my_list[5:])

# elements beginning to end
print(my_list[:])
```

```python
# Index of first occurrence of the value 8

list1= [3, 8, 1, 6, 8, 8, 4]
print(list1.index(8))   # Output: 1
```

## List operations

We can add one item to a list using the <code>append()</code> method or add several items using the <code>extend()</code> method.

```python
# Appending and Extending lists in Python
list1 = [1, 3, 5]
list2 = [2, 4, 6]

list1.append(list2) #we add one element to the list, instead of three elements

print(list1) #we have one new element consisting of a nested list
```

```python
list1 = [1, 3, 5]
list2 = [2, 4, 6]

list1.extend(list2) #we add three elements to the list

print(list1)
```

We can also use + operator to combine two lists. This is also called concatenation. The * operator repeats a list for the given number of times.

```python
list1 = [1, 3, 5]
list2 = [2, 4, 6]

list1 + list2
```

```python
list1 = [1, 3, 5]
list2 = [2, 4, 6]

list1 * 3
```

As lists are mutable, we can change them. For example, we can change the first element as follows:

```python
# Change the element based on the index

A = ["Python", 10, 3.14]
print('Before change:', A)
A[0] = 'Java'
print('After change:', A)
```

```python
#pop() to remove an item at the given index.
A = ["Python", 10, 3.14]

print(A.pop(1)) #remove the second item

print(A)

print(A.pop()) #remove the last item

print(A)
```

```python
#We can use remove() to remove the given item

my_list = ['p','y','t','h','o','n']
my_list.remove('o')

my_list
```

We can also delete an element of a list using the <code>del</code> command:

```python
# Delete the element based on the index

print('Before change:', A)
del(A[0])
print('After change:', A)

# delete the entire list
del A

# Error: List not defined
# print(A)
```

## Copy (copy by reference) the list A

```python
#both A and B are referencing the same list in memory

A = ["Harry Potter", 10, 3.14]
B = A
print('A:', A)
print('B:', B)
```

```python
# if we change value of A[0], B[0] will be changed at the same time
# both A and B are referencing the same list in memory
print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])
```

```python
# Clone (clone by value) the list A
A = ["Harry Potter", 10, 3.14]
B = A[:]
B
```

```python
print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])

#Now if you change A, B will not change
```

## Tuple

<p>A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned. As it is immutable, tuple does not have append(), remove(), pop() functions</p> 

```python
# Empty tuple
my_tuple = ()
print(my_tuple)

# tuple with mixed datatypes
my_tuple = (1, "Hello", 3.4)
print(my_tuple)

# nested tuple
my_tuple = ("Hello", [8, 4, 6], (1, 2, 3))
print(my_tuple)

#my_tuple[0] = ("Henry") #cannot be changed, unlike lists, tuples are immutable

# However, item of mutable element can be changed
my_tuple[1][0] = 9    
print(my_tuple)
```

```python
#only ONE element in tuple?

my_tuple = ("Hello")
print(type(my_tuple))

#in order to make it as tuple, comma should be added
my_tuple = ("Hello",)
print(type(my_tuple))
```

## Tuple Indexing

```python
# Accessing tuple elements using indexing
my_tuple = ('p','y','t','h','o','n')

print(my_tuple[0])   
print(my_tuple[5])
print(my_tuple[-1])
print(my_tuple[-5])
```

```python
# nested tuple
n_tuple = ("Hello", [8, 4, 6], (1, 2, 3))

# nested index
print(n_tuple[0][1])
```

## Tuple Slicing

```python
my_tuple = ('p','y','t','h','o','n')

print(my_tuple[1:4])

print(my_tuple[:-5])

print(my_tuple[4:])

print(my_tuple[:])

```

## Tuple Operations

```python
print((1, 2, 3) + (4, 5, 6))
```

```python
print(("Python",) * 3)
```

```python
# Deleting tuples
my_tuple = ('p','y','t','h','o','n')

#del my_tuple[3] #error

# Can delete an entire tuple
del my_tuple

# NameError: name 'my_tuple' is not defined
print(my_tuple)
```

Methods that add items or remove items are not available with tuple. Only the following two methods are available.

```python
# count() and index() methods
my_tuple = ('p','y','t','h','o','n')

print(my_tuple.count('p')) 
print(my_tuple.index('t'))  
```

```python
#Element-wise sum
x = (1,2,3,4)
y = (3,5,2,1)
z = (2,2,3,1)
print("Original lists:")
print(x)
print(y)
print(z)
print("\nElement-wise sum of the said tuples:")
result = tuple(map(sum, zip(x, y, z)))
print(result)
```

## Membership operators 

```python
x = 'Hello world'
y = ["Apple","Orange","Lemon","Watermelon"]

# Output: True
print('H' in x)

# Output: True
print('hello' not in x)

# Output: True
print("Apple" in y)

# Output: False
print('r' in y)

# Membership test in tuple
my_tuple = ('a', 'p', 'p', 'l', 'e',)

# In operation
print('a' in my_tuple)
print('g' in my_tuple)
```
